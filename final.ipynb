```python
import tkinter as tk

#store history
calculation_history = []

def update_history():
    "Update the history display"
    history_text.delete(1.0,tk.END)
    for item in calculation_history:
        history_text.insert(tk.END, item + "\n")
def clear_fields():
    "Clear all entries"
    name_entry.delete(0, tk.END)
    bought_entry.delete(0, tk.END)
    sold_entry.delete(0, tk.END)
    quantity_entry.delete(0, tk.END)
    quantity_entry.insert(0,"1")
    result_label.config(text="")
def calculate_profit():
    " Calculate profit or loss from player transaction."
    try:
        name = name_entry.get()
        bought_price = float(bought_entry.get())
        sold_price = float(sold_entry.get())
        quantity = int(quantity_entry.get())

#calculate tax
        total_bought = bought_price * quantity
        total_sold = (sold_price * 0.95) * quantity
        profit = total_sold - total_bought

        if profit >= 0:
             result_label.config(text=f"Profit: {profit:,.0f} coins", fg="green")
        else:
            result_label.config(text=f"Loss: {profit:,.0f} coins", fg ="red")
#add history      
        history_entry = f"{name} x{quantity}: {profit:,.0f} coins"
        calculation_history.append(history_entry)
        if len(calculation_history) > 10:
            calculation_history.pop(0)
        update_history()
    
    except: 
        result_label.config(text="Enter valid numbers",fg="red")
        
window = tk.Tk()
window.title("FC 26 Calculator")
window.geometry("400x500")

#format base

calc_frame = tk.Frame(window, padx = 10, pady=10)
calc_frame.pack(fill="x")
tk.Label(calc_frame, text="FC 26 Profit Calculator", font = ("calibri",16, "bold")).pack

#name
tk.Label(calc_frame, text="Player Name:").pack(anchor="w")
name_entry = tk.Entry(calc_frame, width = 30)
name_entry.pack(fill="x", pady = 2)

#purchase price
tk.Label(calc_frame, text="Bought for:").pack(anchor="w")
bought_entry = tk.Entry(calc_frame, width = 30)
bought_entry.pack(fill="x", pady = 2)

#sold
tk.Label(calc_frame, text="Sold for:").pack(anchor="w")
sold_entry = tk.Entry(calc_frame, width=30)
sold_entry.pack(fill="x",pady=2)

#quantity
tk.Label(calc_frame, text="Quantity:").pack(anchor="w")
quantity_entry = tk.Entry(calc_frame, width = 30)
quantity_entry.pack(fill="x", pady = 2)
quantity_entry.insert(0,"1")

#buttons
button_frame = tk.Frame(calc_frame)
button_frame.pack(pady=10)

calc_button = tk.Button(button_frame, text= "Calculate", command = calculate_profit,)
calc_button.pack(side="left", padx=5)

clear_button = tk.Button(button_frame, text="Clear", command=clear_fields)
clear_button.pack(side="left",padx=5)

#display
result_label=tk.Label(calc_frame, text="", font=("calibri", 14, "bold"))
result_label.pack(pady=10)
tk.Label(calc_frame, text= "(Note: 5% EA tax included)", font= ("calibri",7)).pack()

#history format
history_frame = tk.Frame(window, padx=10,pady=10)
history_frame.pack(fill="both", expand=True)
tk.Label(history_frame, text="Last 10 Calculations:" , font=("calibri",12,"bold")).pack()
history_text=tk.Text(history_frame,height=10, width=40)
history_text.pack(fill="both", expand=True , pady=5)

window.mainloop()
```
